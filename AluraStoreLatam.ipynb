### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tiendas = [tienda1, tienda2, tienda3, tienda4]
nombre_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4', ]

tienda1.head()
```

#Análisis de facturación



###Ingreso total por cada tienda

```python
# Calculo el ingreso total de todas las tiendas
# start=1 para igualar el número de índice con el número de tienda

for i, tienda in enumerate(tiendas, start=1):
  total_facturado = tienda['Precio'].sum()
  print(f'Facturación total de Tienda {i}: ${total_facturado:,.2f}')
```

###Ventas por categoría

```python
# Ventas por categoría en Tienda 1

tienda1['Categoría del Producto'].value_counts().reset_index(name='Ventas')
```

```python
# Ventas por categoría en Tienda 2

tienda2['Categoría del Producto'].value_counts().reset_index(name='Ventas')
```

```python
# Ventas por categoría en Tienda 3

tienda3['Categoría del Producto'].value_counts().reset_index(name='Ventas')
```

```python
# Ventas por categoría en Tienda 4

tienda4['Categoría del Producto'].value_counts().reset_index(name='Ventas')
```

### 3. Calificación promedio de la tienda


```python
# Calculo la calificación promedio de cada tienda y muestro los resultados

for i, tienda in enumerate(tiendas, start=1):
  calificacion_promedio = tienda['Calificación'].mean()
  print(f'Calificación promedio de Tienda {i}: {calificacion_promedio:.2f}')
```

### 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos en Tienda 1:

tienda1.groupby("Producto").size().sort_values(ascending=False)
```

```python
# Productos más y menos vendidos en Tienda 2:

tienda2.groupby("Producto").size().sort_values(ascending=False)
```

```python
# Productos más y menos vendidos en Tienda 3:

tienda3.groupby("Producto").size().sort_values(ascending=False)
```

```python
# Productos más y menos vendidos en Tienda 4:

tienda4.groupby("Producto").size().sort_values(ascending=False)
```

###Envío promedio por tienda

```python
for i, tienda in enumerate(tiendas, start=1):
  costo_envio_promedio = tienda['Costo de envío'].mean()
  print(f'Costo promedio de envío en Tienda {i}: ${costo_envio_promedio:.2f}')
```

# Gráficos

### Ingreso total por cada tienda

```python
colores = [
    '#FFB347',  # naranja suave
    '#A8E6CF',  # verde menta (mejor para accesibilidad)
    '#779ECB',  # azul acero pastel
    '#2B547E']   # azul oscuro complementario

# Creo un df auxiliar con totales
df_facturaciones = pd.DataFrame({
    'Tienda': nombre_tiendas,
    'Total': [tienda['Precio'].sum() for tienda in tiendas],
    'Color': colores
})

# Grafico
plt.figure(figsize=(8, 5))
plt.barh(df_facturaciones['Tienda'], df_facturaciones['Total'], color=df_facturaciones['Color'])
plt.gca().invert_yaxis()
plt.xlabel('Total facturado', fontsize=13)
plt.title('Facturación total por tienda', fontsize=16)

plt.show()
```

### Calificación promedio por tienda

```python
# Creo un df auxiliar con promedios
df_calificaciones = pd.DataFrame({
    'Tienda': nombre_tiendas,
    'Calificacion': [tienda['Calificación'].mean() for tienda in tiendas],
    'Color': colores
})

# Grafico
plt.figure(figsize=(8, 5))
plt.bar(df_calificaciones['Tienda'], df_calificaciones['Calificacion'], color=df_calificaciones['Color'])
plt.title('Califiación promedio por tienda', fontsize=16)
plt.xlabel('')
plt.ylabel('Calificación promedio', fontsize=13)
plt.ylim(3.9, 4.1)

for i, calificacion in enumerate(df_calificaciones['Calificacion']):
    plt.text(i, calificacion - 0.010, f'{calificacion:.2f}', ha='center', va='bottom')

plt.show()

```

### Ventas por categoría en cada tienda


```python
# Grafico usando subplots las ventas por categoría. Un gráfico para cada tienda.
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()  # Aplano el array para iterar fácil

# Itero sobre cada tienda y cada subplot
for i, tienda in enumerate(tiendas):
    ax = axs[i]
    ventas_por_categoria = tienda['Categoría del Producto'].value_counts().sort_values()
    ax.barh(ventas_por_categoria.index, ventas_por_categoria.values, color=colores[i])
    ax.set_title(f'{nombre_tiendas[i]}', fontsize=14)
    ax.set_xlabel('Cantidad de ventas')

plt.tight_layout()
plt.show()
```
